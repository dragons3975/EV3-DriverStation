<div class="info">

On va voir ici la boucle while qui répète les instructions tant qu'une condition est vraie avec la syntaxe:

</div>

```python
    while condition:
        instructions à répéter tant que la condition est vraie
```	

<div class="info">

Par exemple pour compter les nombres de 1 à 10

</div>

```python
    compteur = 1
    while compteur <= 10:
        print("La valeur de compteur est: ", compteur)
        compteur = compteur + 1
```	

Affiche tous les nombres de 10 à 20 avec une boucle while.

Décompte les nombres de 10 à 0. Lorsque tu atteint 0 affiche "Décollage !"

Affiche tous les nombres impairs de 11 à 21

Code la fonction somme(n) qui calcule la some: $1 + 2 + 3 + ... + n$. Quelle est la valeur de $1 + 2 + 3 + ... + 99 + 100$?

**Bonus** : Code la fonction ``approximate_pi()`` qui approxime $\pi$ avec la formule: 
$\pi = 2 \times \Big( \dfrac{1\times2}{1\times3} + \dfrac{1\times2\times3}{1\times3\times5} + \dfrac{1\times2\times3\times4}
{1\times3\times5\times7} + ... \Big)$

Utilise une boucle et 100 itération pour approximer $\pi$.
