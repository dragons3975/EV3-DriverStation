Les boucles permettent de répéter une série d'instruction un certain nombre de fois ou jusqu'à ce qu'une condition soit remplie.

<div class="info">

Il existe deux types de boucles en Python (et dans la majorité des langages de programmations):

<b>Les boucles for</b> permettent de répéter une série d'instruction n fois. Elle s'utilise avec <b>for i in range(n):</b>.

<b>Les boucles while</b> permettent de répéter une série d'instruction tant qu'une condition est valide. Elle s'utilise avec <b>while condition:</b>.

</div>

Execute le code suivant. Que fait ce code? Dans quel ordre sont executés les lignes de code?

````python
for age in range(25):
    majeur = age >= 18
    if age == 0:
        print("Aujourd'hui tu es né!")
    elif majeur:
        print("Aujourd'hui tu as ", age, "ans. Tu es majeur.")
    else:
        print("Aujourd'hui tu as ", age, "ans. Tu n'es pas majeur.")
````
