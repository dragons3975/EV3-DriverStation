- `robot.avance(temps, puissance)` : le robot avance pendant `temps` secondes avec une puissance de `puissance` (entre 0 et 1).

- `robot.tourne_gauche(temps, puissance)` : le robot tourne à gauche pendant `temps` secondes avec une puissance de `puissance` (entre 0 et 1).
- `robot.tourne_droite(temps, puissance)` : le robot tourne à droite pendant `temps` secondes avec une puissance de `puissance` (entre 0 et 1).
